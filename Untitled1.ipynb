```python
import pandas as pd
import numpy as np
import xlrd
import os 

path = './'

read_csv = os.path.join(path, "ENB2012_data.xlsx")
df = pd.read_excel(read_csv)

df
```

```python
df.columns = ['Rel Compacteness', 'Surface Area', 'Wall Area', 'Roof Area', 'Overall Height', 'Orientation', 'Glazing Area', 'Glazing Area Distribution', 'Heating Load', 'Cooling Load']
df
```

```python
df = df.reindex(np.random.permutation(df.index))  # reindexing the data set to avoid the overfitting 
df
```

```python
df = df.reindex(np.random.permutation(df.index))
df.reset_index(inplace=True, drop=True)
display(df)
```

```python
from scipy.stats import zscore

x_columns = df.columns
for col in x_columns:
    df[col] = zscore(df[col])
    
x = df[x_columns].values
    
df.head()
```

```python
X = df[['Rel Compacteness', 'Surface Area', 'Wall Area', 'Roof Area', 'Overall Height', 'Orientation', 'Glazing Area', 'Glazing Area Distribution']].values
y1 = df[['Heating Load']].values
y2 = df[['Cooling Load']].values
```

```python
import pandas as pd
import os
import numpy as np
from sklearn import metrics
from scipy.stats import zscore
from sklearn.model_selection import KFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

from sklearn.model_selection import train_test_split

# Cross-Validate
kf = KFold(5, shuffle=True, random_state=42) 
    
oos_y1 = []
oos_pred1 = []

oos_y2 = []
oos_pred2 = []

fold = 0

for train, test in kf.split(X):
    fold+=1
    print(f"Fold #{fold}")
        
    x1_train = X[train]
    y1_train = y1[train]
    x1_test = X[test]
    y1_test = y1[test]
    
    
    x2_train = X[train]
    y2_train = y2[train]
    x2_test = X[test]
    y2_test = y2[test]
    
 
    model = Sequential()
    model.add(Dense(20, input_dim=8, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    #model.fit(x1_train,y1_train,validation_data = (x1_test,y1_test),verbose=0,epochs=500)
    #model.fit(x2_train,y2_train,validation_data = (x2_test,y2_test),verbose=0,epochs=500)
 
    x1_train, x1_test, y1_train, y1_test = train_test_split(X, y1, test_size = 0.25, random_state = 0)
    x2_train, x2_test, y2_train, y2_test = train_test_split(X, y2, test_size = 0.25, random_state = 0)

    pred1 = model.predict(x1_test)
    pred2 = model.predict(x2_test)
    
    oos_y1.append(y1_test)
    oos_y2.append(y2_test)
    
    oos_pred1.append(pred1)    
    oos_pred2.append(pred2) 

    # Measure this fold's RMSE
    score = np.sqrt(metrics.mean_squared_error(pred1,y1_test))
    print(f"Fold score for heating (RMSE): {score}")
    
    score = np.sqrt(metrics.mean_squared_error(pred2,y2_test))
    print(f"Fold score for cooling (RMSE): {score}")
    

# Build the oos prediction list and calculate the error.
oos_y1 = np.concatenate(oos_y1)
oos_y2 = np.concatenate(oos_y2)

oos_pred1 = np.concatenate(oos_pred1)
oos_pred2 = np.concatenate(oos_pred2)

score1 = np.sqrt(metrics.mean_squared_error(oos_pred1,oos_y1))
print(f"Final, out of sample score for heating (RMSE): {score1}") 

score2 = np.sqrt(metrics.mean_squared_error(oos_pred2,oos_y2))
print(f"Final, out of sample score for cooling (RMSE): {score2}")    


    
# Write the cross-validated prediction
oos_y1 = pd.DataFrame(oos_y1)
oos_y2 = pd.DataFrame(oos_y2)

oos_pred1 = pd.DataFrame(oos_pred1)
oos_pred2 = pd.DataFrame(oos_pred2)

oosDF = pd.concat( [df, oos_y1, oos_pred1, oos_y2, oos_pred2],axis=1 )

print(oosDF)

#oosDF.to_csv(filename_write,index=False)
```
